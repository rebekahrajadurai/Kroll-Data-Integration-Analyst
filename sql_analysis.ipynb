## Count of devices that do not have a version number:

```python
%%sql
WITH cte_version_number AS (
	SELECT 
		model_number, 
		CASE WHEN version_number = 0 THEN 'N/A' ELSE version_number END AS version_number, 
		COUNT(*) AS VersionNumCount FROM device
	GROUP BY model_number, version_number
	ORDER BY version_number
)
SELECT COUNT(*) AS DevicesWithNoVersionNumber
FROM cte_version_number
WHERE version_number = 'N/A';
```

This information can be useful for businesses that need to track or report on the status of their devices. A company that provides support for devices may need to know how many devices are running without a version number to ensure that they are up-to-date and secure. The absence of a version number may indicate that a device is an older model, and may have been discontinued or replaced by a newer version. Sales and customer support teams can use this information to provide accurate information to customers who may be interested in purchasing or servicing these devices.


# Number of Non-Sterile Devices by Brand with No Storage/Handling Information

```python
%%sql
SELECT 
  company_name,
  brand_name, 
  COUNT(brand_name) AS brand_name_count
FROM device d
LEFT JOIN device_environment_conditions denvc 
    ON d.id = denvc.device_id
WHERE d.id NOT IN 
    (
        SELECT dt.device_id
        FROM device_terms dt
        WHERE dt.name NOT LIKE '%non-sterile%'
    )
    AND denvc.storage_handling_type IS NULL
GROUP BY brand_name
ORDER BY brand_name_count DESC;
```

A company could use this information to assess its own position within the market relative to its competitors. They could also use this information to identify potential areas for investment or development in terms of new devices that address gaps in the market or improve upon existing products. This can also help evaluate areas for improvement or alternative solutions.

## Device Count and RX/OTC Breakdown by Product Code Name











```python
%%sql
SELECT 
	dpc.product_code_name,
	SUM(d.is_rx) AS RX, 
	SUM(d.is_otc) AS OTC,
	COUNT(*) AS Total_product_code_name
FROM device_product_codes dpc 
JOIN device d 
	ON dpc.device_id = d.id
GROUP BY product_code_name 
ORDER BY COUNT(*) DESC;
```

Each product category either is available only as presciption or only over-the-counter. Product management teams can use this to identify which product code names (categories) have the most prescription and OTC devices. This  can help companies in making strategic decisions on the products to produce and the distribution channels to use. ## Average Number of Updates By Company Per Year


```python
%%sql
-- looking at average updates per year by company
-- for their device offerings, to evaluate the need for a new product on the market
WITH cte_publish_data AS (
	SELECT 
		company_name, 
		year(publish_date) AS publish_year, 
		month(publish_date) AS publish_month, 
		COUNT(*) as publish_count
	FROM device
	GROUP BY company_name, year(publish_date), month(publish_date)
), 
cte_publish_years AS (
	SELECT 
		company_name,
		publish_year,
		SUM(publish_count) AS updates_year
	FROM cte_publish_data
	GROUP BY company_name, publish_year
)
SELECT 
	company_name, 
	ROUND(SUM(updates_year) / COUNT(*), 0) AS avg_updates_per_year
FROM cte_publish_years 
GROUP BY company_name
ORDER BY avg_updates_per_year DESC;
```

This shows that Gold Standard Orthopedics is the biggest competitor for this company that is trying to create a new product. This company has all of its brands with high quality products having many updates per year to keep its products up to date. Companies that release a higher number of updates per year may be more dedicated to improving their products and keeping up with the latest technologies, which can be an advantage in a competitive market.
The average number of updates per year for each company can help product managers and investors to evaluate the company's product development cycle and the rate of innovation compared to the competition.
If a company's average number of updates per year is significantly lower than its competitors, it may indicate that the company is struggling to keep up with the market demands and needs to invest more in its product development and innovation.

## Frequency of updates made to products in each category


```python
%%sql 
SELECT 	
	company_name, 
  dpc.product_code_name, 
	year(publish_date) AS Year, 
  MONTH(publish_date) AS Month,
	COUNT(*) OVER (PARTITION BY YEAR(publish_date), MONTH(publish_date)) AS update_frequency_count
FROM device d 
JOIN device_product_codes dpc 
	ON dpc.device_id = d.id 
GROUP BY dpc.product_code_name, YEAR(publish_date), MONTH(publish_date);
```

## Saturation of products with prescription vs OTC

```python
%%sql
SELECT 
 dpc.product_code_name, 
    CASE WHEN d.is_rx = 1 THEN 'Prescription Required' ELSE 'No prescription required' END AS is_prescription_required, 
    CASE WHEN d.is_otc = 1 THEN 'Available over the counter' ELSE 'Not available over the counter' END AS is_over_the_counter, 
    COUNT(*) 
FROM device d
JOIN device_product_codes dpc on d.id = dpc.device_id
GROUP BY dpc.product_code_name, d.is_rx, d.is_otc
ORDER BY COUNT(*) DESC;
```

The resulting table shows Spinal cord devices in the top 3. With spinal inter laminal appliance at the highest with 213 products, and spinal pedicle fixation orthosis with 107 products. They both obviously are prescription only products. This shows that understand the spinal devices are in a highly competitive landscape of available products, especially with the the level of regulation(where earlier we saw that they highly produced as sterile products with proper instructions, and high update frequency. This indicates that it has a high market demand. 

## View of Full Device Data

```python
%%sql
CREATE VIEW vw_full_device_data
AS
SELECT
 d.brand_name, d.version_status, d.record_status, d.version_number, d.version_date, d.publish_date, d.distribution_end_date, d.distribution_status, d.model_number,
    d.company_name, d.device_description, d.is_rx, d.is_otc,
    dc.email_address, dc.phone_number,
    denvc.storage_handling_type, storage_handling_special_condition_text,
    di.device_identifier, di.issuing_agency, dms. submission_number, dms.supplement_number,
    dpc.product_code, dpc.product_code_name,
    ds.size_type, ds.size, ds.size_unit, ds.size_text,
    dst.is_device_sterile,
    dst.require_initixal_sterilization,
    dst.method,
    dt.name, dt.pt_definition
FROM device d
LEFT JOIN device_contacts dc 
	ON dc.device_id = d.id
LEFT JOIN device_environment_conditions denvc 
	ON denvc.device_id = d.id
LEFT JOIN device_identifiers di 
	ON di.device_id = d.id
LEFT JOIN device_market_submissions dms 
	ON dms.device_id = d.id
LEFT JOIN device_product_codes dpc 
	ON dpc.device_id = d.id
LEFT JOIN device_sizes ds on ds.device_id = d.id
LEFT JOIN device_sterilization dst 	
	ON dst.device_id = d.id
LEFT JOIN device_terms dt
	ON dt.device_id = d.id;
```

```python
%%sql
SELECT *
FROM vw_full_device_data
LIMIT 10;
```

This view can help the company gain a comprehensive understanding of the medical devices on the market and monitor their performance across various dimensions such as size, sterilization, product codes, market submissions, etc. This view can also be used to track changes in the version number, status and date of a device, which can help them ensure regulatory compliance and manage recalls if necessary. Sales teams can use this view to identify devices that are available for sale, their distribution status and the contact information for the manufacturer, which can help them manage relationships with clients and customers.
