Importing necessary libraries

To make a web page request
and to interact with an SQL database

```python
from bs4 import BeautifulSoup
import json
import requests
from sqlalchemy import create_engine
```

```python
web_url = "https://accessgudid.nlm.nih.gov/api/v2/devices/history.json"
```

setting a dictionary of parameters to pass to the URL

```python
params = {"di": "03596010535344"}
```

making a GET request with the defined URL and the dict of parameters

```python
response = requests.get(web_url, params=params)
```
